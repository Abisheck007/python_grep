```python
import pandas as pd

india_weather = pd.DataFrame({
    'date': ['2023-01-01', '2023-01-02', '2023-01-03'],
    'temperature': [25, 26, 27],
    'humidity': [60, 65, 70],
    'wind_speed': [10, 12, 14]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    'date':['2023-01-01', '2023-01-02', '2023-01-03'],
    'temperature': [30, 31, 29],
    'humidity': [55, 50, 52],
    'wind_speed': [8, 9, 7]
})

us_weather
```

```python
df  = pd.concat([india_weather, us_weather])
df
```

```python
##now see above the index is unoder ed, we can reset the index
df = pd.concat([india_weather,us_weather], ignore_index=True)
df
```

```python
##we can also give keys to the dataframes
df = pd.concat([india_weather, us_weather], keys=['India', 'US'])
df
```

```python
df.loc['India']
```

```python
addcol = pd.Series(['Humid','dry','cold'], name='weather_condition')
addcol
```

```python
newdata = pd.concat([india_weather, addcol], axis=1)
newdata
```
