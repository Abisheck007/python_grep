```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
sea = sns.load_dataset('car_crashes')
sea.head(5)
```

```python

#heatmaps:it used to bring columns and rows together to see the correlation between them.
sea_corr = sea.corr(numeric_only=True)
sns.heatmap(sea_corr, annot=True, cmap="coolwarm") 

```

```python
#alternatively we can use pivot table to create a heatmap
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
flights_pivot
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
species = iris.pop('species')
sns.clustermap(iris)
```

```python
#pair grid
abi = sns.PairGrid(iris, hue='species')
abi.map(plt.scatter)
```

```python
#facetgrid
tips_df = sns.load_dataset('tips')
tips = sns.FacetGrid(tips_df,col='time',hue='smoker')
tips.map(plt.scatter,'total_bill','smoker')
```

```python
#regression plots
sns.lmplot(x='total_bill', y='tip',hue='sex', data=tips_df, markers=['v', '^'])
```
