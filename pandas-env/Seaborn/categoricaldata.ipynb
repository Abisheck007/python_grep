```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
#bar plot along with estimator
sns.barplot(x = 'sex' , y='total_bill', data=tips_df,estimator=np.mean)#IN Estimator we can use any function like np.mean, np.sum, np.median ,np.std,np.var etc
```

```python
#count plot :used to count the no of occurrences of each category
sns.countplot(x = 'sex',data=tips_df)
```

```python
#box plot is used to show the distribution of data based on give x y and hue values
sns.boxplot(x='day',y = 'total_bill',data=tips_df,hue='sex')
```

```python
#violin plot is used to show the distribution of data based on give x y and hue values
#violin plot is the combination of box plot and kde plot
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

```python
#strip plt:gives us the scactter plottings and one variable is categorical
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex',dodge=True)
```

```python
#swarm plts
sns.swarmplot(x='day',y='total_bill',data=tips_df,hue = 'sex',dodge=True)
```
