```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('ComputerSales.csv')
df.head(5)
```

```python
#we dont use pandas to get the dataset instead we use seaborn inbuilt dataset
#first we see the dataset names
sea = sns.get_dataset_names()
sea
```

```python
#the above shown are the inbuilt datasets in seaborn here we use one of them here
sea = sns.load_dataset('car_crashes')
sea.head(5)
```

```python
#we take seperate data from the dataset and form a garph using distribution plot
sns.displot(sea['alcohol'],kde=True)
```

```python
#joint plot is used to plot the distribution of two variables
abi = sns.jointplot(data = sea, x = 'alcohol', y = 'speeding', kind = 'hex')#in kind we can give kde,regression,scatter,hex
abi
```

```python
#KDE Plot (kernel density estimation) is used to visualize the distribution of a variable
sns.kdeplot(data = sea, x = 'alcohol', y = 'speeding', fill=True)
```

```python
#KDE Plot (kernel density estimation) is used to visualize the distribution of a variable
sns.kdeplot(sea['alcohol'],  color='r',fill=True)
```

```python
#Pair plots
sns.pairplot(sea)
```

```python
#another pair plot ex using tips
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df , hue='sex',palette='coolwarm')
```

```python
#rug plot:used to plot single column of data points in dataframes
sns.rugplot(sea['alcohol'])
```

```python
#styling 
# sns.set_style('darkgrid')
# sns.set_style('whitegrid')
sns.set_style('dark')
sns.jointplot(data = sea, x = 'alcohol', y = 'speeding', kind = 'reg')#in kind we can give kde,regression,scatter,hex
```
