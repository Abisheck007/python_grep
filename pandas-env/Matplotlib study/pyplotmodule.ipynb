```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,6])
y = np.array([0,250])
plt.plot(x, y,color='red', linewidth=2, label='Line 1')
plt.scatter(x, y, color='red')
plt.show()
```

```python
##we also do poiting without using line and scatter 
x = np.array([0,6])
y = np.array([0,250])
plt.plot(x, y,'o',color='green')

plt.show()
```

```python
x = np.array([0,1,3,5,6])
y = np.array([0,5,8,2,5])
plt.plot(x, y,color='red', linewidth=2, label='Line 1')
plt.scatter(x, y, color='red')
plt.show()
```

```python
y = np.array([3,7,5,2])
plt.plot(y)

plt.show()
```

```python
#marker
x = np.array([0,1,2,3,4,5])
plt.plot(x,marker='o')
plt.show()
```

```python
#shortcuts
x = np.array([0,1,2,3,4,5])
plt.plot(x ,'o:b')#it mean plot points with dotted lines and with blue color this is what it means  like this : we can use three more -,--,-.
plt.show()
```

```python
  #marker size
x = np.array([0,1,2,3,4,5])
plt.plot(x ,'o:b',ms=20)#it means marker size is 20 similarly like changing the width
plt.show()
```

```python
#marker edge color
x = np.array([0,1,2,3,4,5])
plt.plot(x ,'o:b' ,mec='red')#it means marker edge color is red
plt.show()
```

```python
#line style
abi = np.array([0,1,2,3,4,5])
plt.plot(abi, linestyle='--')#it means dashed line
plt.show()
```

```python
#line width
abi = np.array([0,1,2,3,4,5])
plt.plot(abi, linestyle='--', lw=5)  # it means line width is 5
plt.show()
```

```python
#multi line ploting
x1 = np.array([0,1.5,2.7,4,5])
x2 = np.array([0,5,2.5,5,0])


y1 = np.array([1,2.5,3.5,1.7])
y2 = np.array([0,5,0,3])
plt.plot(x1, x2, 'o-b', ms=4, label='Line 1' ,linewidth=10)
plt.plot(y1, y2, 'o-r',linewidth=10 )

plt.show()
```

```python
#labels
x = np.array([0,1,2,3,4,5])
y = np.array([0,5,8,2,5,1])
plt.plot(x, y)
plt.xlabel('X-axis Label')  # Set the x-axis label
plt.ylabel('Y-axis Label')  # Set the y-axis label
plt.title('Plot Title')  # Set the title of the plot
plt.show()
```

```python
#grid
x = np.array([0,1,2,3,4,5])
y = np.array([0,5,8,2,5,1])
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#we can also use grid for single axis
x = np.array([0,1,2,3,4,5])
y = np.array([0,5,8,2,5,1])
plt.plot(x, y)
# plt.grid(axis='x') 
# Only grid for the x-axis
plt.grid(axis='y')  # Only grid for the y-axis
plt.show()
```

```python
#we can  also do styles in grid
x = np.array([0,1,2,3,4,5])
y = np.array([0,5,8,2,5,1])
plt.plot(x, y)
plt.grid(color='red', linestyle='--')  # Custom grid style
plt.show()
```

```python
#subplot:using subplot we can plot multiple graphs in a single box or figure
x = np.array([0,1,2,3])
y = np.array([0,5,8,2])
plt.subplot(3,1,1)
plt.plot(x,y)
plt.title('Subplot 1')

x1 = np.array([0,2,4,6])
y1 = np.array([0,3,5,1])
plt.subplot(3,1,2)
plt.plot(x1,y1)
plt.title('Subplot 2')

x2 = np.array([0,1,3,5])
y2 = np.array([0,5,8,2])
plt.subplot(3,1,3)
plt.plot(x2,y2)
plt.title('Subplot 3')
plt.suptitle('Multiple Subplots Example')  # Title for the entire figure
plt.show()
```

```python
#scatter plots
x = np.array([0,1,2,3,4,5])
y = np.array([0,5,8,2,5,1])
plt.scatter(x, y, color='blue', marker='o', s=100)  # s is the size of the markers
plt.show()
```

```python
##multiple scatter plots
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,5,8,2,5,1])
x2 = np.array([0,1,2,3,4,5])
y2 = np.array([1,4,6,3,2,5])
plt.scatter(x1,y1,color="blue")
plt.scatter(x2,y2,color="hotpink")
plt.show()
```

```python
#color bar in scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)  # Random colors for each point
plt.scatter(x, y, c=colors, cmap='viridis', s=100)  # s is the size of the markers
plt.colorbar()  # Show color bar
plt.title('Scatter Plot with Color Bar')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

```python
#bars
y= np.array([10,20,30,40,50])
x = np.array([1,2,3,4,5])
plt.bar(x, y, color='blue', width=0.5)  # width is the width of the bars
plt.show()
```

```python
#horizontal bars
y= np.array([10,20,30,40,50])
x = np.array([1,2,3,4,5])
plt.barh(x, y)  # width is the width of the bars
plt.show()
```

```python
#pie charts
x = np.array([10,20,30,40])
mylabels = ['A', 'B', 'C', 'D']
plt.pie(x, labels=mylabels, autopct='%1.1f%%')  # autopct shows percentage
```

```python
#explode and legent method // explode is used to seperate one single pie from the rest
x = np.array([10,20,30,40])
mylabels = ['A', 'B', 'C', 'D']
myexplode = (0.2, 0, 0, 0)  # only "explode" the first slice
plt.pie(x, labels=mylabels, autopct='%1.1f%%', explode=myexplode,shadow=True)
plt.legend(title='percentage analysis')  # Add a legend with a title
plt.show()
```
