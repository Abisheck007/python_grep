```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,6])
y = np.array([0,250])
plt.plot(x, y,color='red', linewidth=2, label='Line 1')
plt.scatter(x, y, color='red')
plt.show()
```

```python
##we also do poiting without using line and scatter 
x = np.array([0,6])
y = np.array([0,250])
plt.plot(x, y,'o',color='green')

plt.show()
```

```python
x = np.array([0,1,3,5,6])
y = np.array([0,5,8,2,5])
plt.plot(x, y,color='red', linewidth=2, label='Line 1')
plt.scatter(x, y, color='red')
plt.show()
```

```python
y = np.array([3,7,5,2])
plt.plot(y)

plt.show()
```

```python
#marker
x = np.array([0,1,2,3,4,5])
plt.plot(x,marker='o')
plt.show()
```

```python
#shortcuts
x = np.array([0,1,2,3,4,5])
plt.plot(x ,'o:b')#it mean plot points with dotted lines and with blue color this is what it means  like this : we can use three more -,--,-.
plt.show()
```

```python
  #marker size
x = np.array([0,1,2,3,4,5])
plt.plot(x ,'o:b',ms=20)#it means marker size is 20 similarly like changing the width
plt.show()
```

```python
#marker edge color
x = np.array([0,1,2,3,4,5])
plt.plot(x ,'o:b' ,mec='red')#it means marker edge color is red
plt.show()
```

```python
#line style
abi = np.array([0,1,2,3,4,5])
plt.plot(abi, linestyle='--')#it means dashed line
plt.show()
```

```python
#line width
abi = np.array([0,1,2,3,4,5])
plt.plot(abi, linestyle='--', lw=5)  # it means line width is 5
plt.show()
```

```python
#multi line ploting
x1 = np.array([0,1.5,2.7,4,5])
x2 = np.array([0,5,2.5,5,0])


y1 = np.array([1,2.5,3.5,1.7])
y2 = np.array([0,5,0,3])
plt.plot(x1, x2, 'o-b', ms=4, label='Line 1' ,linewidth=10)
plt.plot(y1, y2, 'o-r',linewidth=10 )

plt.show()
```

```python
#labels
x = np.array([0,1,2,3,4,5])
y = np.array([0,5,8,2,5,1])
plt.plot(x, y)
plt.xlabel('X-axis Label')  # Set the x-axis label
plt.ylabel('Y-axis Label')  # Set the y-axis label
plt.title('Plot Title')  # Set the title of the plot
plt.show()
```
