```python
import numpy as np
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt

abi = np.array([[2,4],[3,4],[5,6],[4,5]])
simplices = Delaunay(abi).simplices
plt.triplot(abi[:,0], abi[:,1], simplices,color='green', linewidth=1.5)
plt.scatter(abi[:,0], abi[:,1], color='red')
plt.show()
```

```python
#using convex hull
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt

abi = np.array([[2,4],[3,4],[3,0],[2,2],[4,1],[1,2],[5,0],[3,1],[1,2],[0,2]])
hull = ConvexHull(abi)
hull_point = hull.simplices

plt.scatter(abi[:,0], abi[:,1], color='red')
for i in hull_point:
    plt.triplot(abi[:,0], abi[:,1],color='green', linewidth=1.5)
plt.show()
```

```python
from scipy.spatial import KDTree#take neighbour queries

abi = [(1,-1),(2,3),(-2,3),(2,-3)]
tree = KDTree(abi)
query = tree.query((1,1))
query
```

```python
#distance method nothing but euclidean distance

from scipy.spatial.distance import euclidean
p1 = (1,0)
p2 = (10,2)
distance = euclidean(p1, p2)
distance

```
